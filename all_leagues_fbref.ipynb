# Fixing the special character problem in player names

In order to join our transfermarkt and fbref data, we need to ensure that the 'Player' column in both dataframes contains matching information. Therefore we will clean the names by removing non-English characters from both dataframes.

```python
#Creating a list of special characters found in player names. These are characters that are common in European languages/names but not in English.

fbref_player_list = [a for a in fbref_df['Player']]
tm_player_list = [a for a in transfermarkt['Player']]
full_player_list = fbref_player_list + tm_player_list

character_list = []

for a in full_player_list:
    for b in a:
        if b not in character_list:
            character_list.append(b)

            
character_list.sort()


#Only selecting non-traditional characters that are found in certain languages
special_char = character_list[56:]
```

```python
#Identifying the non-traditional characters in our Player names.
len(special_char)
```

```python
#Manually created list of replacements for our special character list above. The replacements are arranged in the same order as their respective original special characters.
replacement = ['','A','C','E','N','O','O','O','U','ss','a',
               'a','a','a','a','a','ae','c','e','e','e',
               'e','i','i','i','d','n','o','o','o','o',
               'o','u','u','y','a','a','C','c','c','D',
               'e','e','g','I','i','L','l','n','o','r',
               's','S','s','z','S','s','t']
```

```python
#Replacing Special characters in the 'Player' Column of the fbref dataframe.

fbref_clean_players = []
for a in fbref_df['Player']:

    string = a
        
    player_string = []
    for z in string:
        if z in special_char:
            for e in list(range(len(special_char))):
                if z == special_char[e]:
                    player_string.append(replacement[e])
                else:
                    pass
        else:
            player_string.append(z)
    player_string = ''.join(player_string)
    fbref_clean_players.append(player_string)

        
            
fbref_df['Player'] = fbref_clean_players             
```

```python
#Observing if the 'Player' column has been changed in the fbref dataframe.
fbref_df
```

```python
#Replacing Special characters in the 'Player' Column of the transfermarkt dataframe.

tm_clean_players = []
for a in transfermarkt['Player']:

    string = a
        
    player_string = []
    for z in string:
        if z in special_char:
            for e in list(range(len(special_char))):
                if z == special_char[e]:
                    player_string.append(replacement[e])
                else:
                    pass
        else:
            player_string.append(z)
    player_string = ''.join(player_string)
    tm_clean_players.append(player_string)
    

transfermarkt['Player'] = tm_clean_players
```

```python
#Observing if the 'Player' column has been changed in the transfermarkt dataframe.
transfermarkt
```

```python
#Combining the transfermarkt and fbref on the common 'Player' column.

final_df = pd.merge(transfermarkt, fbref_df, on='Player')
```

# Fixing our Missing Player Problem

## Identifying missing players from the <i>fbref</i> dataset

```python
#Comparing our fbref dataframe with the final_df dataframe to find missing players

df1 = final_df[['Player','Squad (20-21)']]
df2 = fbref_df[['Player','Squad (20-21)']]

check_df_fbref = pd.merge(df1, df2, how='outer', suffixes=('','_y'), indicator=True)
failed_fbref = check_df_fbref[check_df_fbref['_merge']=='right_only'][['Player','Squad (20-21)']].sort_values('Squad (20-21)')
```

```python
#Comparing our transfermarkt dataframe with the final_df dataframe to find missing players

df1 = final_df[['Player','Club','Value']]
df2 = transfermarkt[['Player','Club','Value']]

check_df_tm = pd.merge(df1, df2, how='outer', suffixes=('','_y'), indicator=True)
failed_tm = check_df_tm[check_df_tm['_merge']=='right_only'][['Player','Club','Value']].sort_values('Value',ascending=False)
```
