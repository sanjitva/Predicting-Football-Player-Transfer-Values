```python
#Creating a list of special characters found in player names. These are characters that are common in European languages/names but not in English.

fbref_player_list = [a for a in final['Player']]
tm_player_list = [a for a in tm['Player']]
full_player_list = fbref_player_list + tm_player_list

character_list = []

for a in full_player_list:
    for b in a:
        if b not in character_list:
            character_list.append(b)

            
character_list.sort()


#Only selecting non-traditional characters that are found in certain languages
special_char = character_list[56:]
```

```python
#Manually created list of replacements for our special character list above. The replacements are arranged in the same order as their respective original special characters.
replacement = ['','A','A','C','E','N','O','O','O','U','ss',
                'a','a','a','a','a','a','ae','c','e','e','e',
                'e','i','i','i','d','n','o','o','o','o','o',
                'u','u','y','a','a','C','c','C','c','D','d',
                'e','e','g','I','i','l','L','l','n','n','o',
                'r','S','s','S','s','u','Z','z','S','s','t']


#Replacing Special characters in the 'Player' Column of the fbref dataframe.

fbref_clean_players = []
for a in final['Player']:

    string = a
        
    player_string = []
    for z in string:
        if z in special_char:
            for e in list(range(len(special_char))):
                if z == special_char[e]:
                    player_string.append(replacement[e])
                else:
                    pass
        else:
            player_string.append(z)
    player_string = ''.join(player_string)
    fbref_clean_players.append(player_string)

        
            
final['Player'] = fbref_clean_players  


#Replacing Special characters in the 'Player' Column of the transfermarkt dataframe.

tm_clean_players = []
for a in tm['Player']:

    string = a
        
    player_string = []
    for z in string:
        if z in special_char:
            for e in list(range(len(special_char))):
                if z == special_char[e]:
                    player_string.append(replacement[e])
                else:
                    pass
        else:
            player_string.append(z)
    player_string = ''.join(player_string)
    tm_clean_players.append(player_string)
    

tm['Player'] = tm_clean_players
```
